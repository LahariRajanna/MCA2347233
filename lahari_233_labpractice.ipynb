```python
import tkinter as tk
from tkinter import ttk

root = tk.Tk()
root.title("Railway Management System")

# Create the form
form = ttk.Frame(root)
form.grid(row=0, column=0, padx=10, pady=10)

# Add form fields
passenger_label = ttk.Label(form, text="Passenger Name:")
passenger_label.grid(row=0, column=0, sticky=tk.W)
passenger_entry = ttk.Entry(form)
passenger_entry.grid(row=0, column=1)

age_label = ttk.Label(form, text="Age:")
age_label.grid(row=1, column=0, sticky=tk.W)
age_entry = ttk.Entry(form)
age_entry.grid(row=1, column=1)


# Add radio buttons
gender_label = ttk.Label(form, text="Gender:")
gender_label.grid(row=3, column=0, sticky=tk.W)
gender_var = tk.StringVar()
male_radio = ttk.Radiobutton(form, text="Male", variable=gender_var, value="Male")
male_radio.grid(row=3, column=1)
female_radio = ttk.Radiobutton(form, text="Female", variable=gender_var, value="Female")
female_radio.grid(row=3, column=2)

# Add menu
menu_label = ttk.Label(form, text="Select City:")
menu_label.grid(row=4, column=0, sticky=tk.W)
menu_var = tk.StringVar()
menu_var.set("Select")
menu_options = ['Select',"Bangalore", "Mumbai", "Delhi", "Chennai"]
menu_dropdown = ttk.OptionMenu(form, menu_var, *menu_options)
menu_dropdown.grid(row=4, column=1)

# Add checkbox
#checkbox_var = tk.IntVar()
#checkbox = ttk.Checkbutton(form, text="I agree to the terms and conditions", variable=checkbox_var)
#checkbox.grid(row=2, columnspan=2)

# Add buttons
submit_button = ttk.Button(form, text="Submit")
submit_button.grid(row=5, columnspan=3)

root.mainloop()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Line graph
x = np.array([121, 155, 233, 365, 248])
y = np.array([10, 12, 17, 22, 20])
plt.plot(x, y)
plt.xlabel('Distance(kms)')
plt.ylabel('Time(hrs)')
plt.title('Time-Distance Line Graph')
plt.show()

# Bar graph
x = np.array([2017, 2018, 2019, 2020, 2021])
y = np.array([8.7, 12.3, 24.1, 7.2, 2.76]) 
plt.bar(x, y)
plt.xlabel('No. of passengers(lakhs)')
plt.ylabel('Year')
plt.title('Passenger Traddic-Bar Graph')
plt.show()

# Scatter plot
x = np.array([12, 27, 18, 34, 82]) 
y = np.array([2017, 2018, 2019, 2020, 2021])
plt.scatter(x, y)
plt.xlabel('No. of delayed Train')
plt.ylabel('Year')
plt.title('Scatter Plot')
plt.show()

# Correlation graph or heat map
data = np.random.rand(5, 5)
plt.imshow(data)
plt.colorbar()
plt.title('Correlation Graph or Heat Map')
plt.show()
```
