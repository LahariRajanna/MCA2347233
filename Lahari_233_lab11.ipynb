```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset 
df = pd.read_csv('rail.csv')

# Explore the data
print("Sample data:")
print(df.head())


# Data Visualization

# Line Plot (Speed vs. Distance)
plt.figure(figsize=(10, 6))
plt.plot(df['Distance'], df['Speed'], marker='o', linestyle='-', color='b')
plt.title('Line Plot: Speed vs. Distance')
plt.xlabel('Distance')
plt.ylabel('Speed (km/h)')
plt.grid(True)
plt.tight_layout()

# Scatter Plot (Fare vs. Distance)
plt.figure(figsize=(8, 6))
plt.scatter(df['Distance'], df['Travel_Time'], alpha=0.5)
plt.title('Scatter Plot: Travel Time vs. Distance')
plt.xlabel('Distance')
plt.ylabel('Travel Time (minutes)')
plt.grid(True)
plt.tight_layout()


# Show the plot
plt.show()
```
